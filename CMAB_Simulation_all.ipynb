```python
# plotting inline
%matplotlib inline

# working directory
#import os; os.chdir('/home/gdmarmerola/ts_demo')

# importing necessary modules
import sys, random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy.stats import beta as beta_dist
from scipy.stats import norm as norm_dist
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from tqdm import tqdm
from sklearn.linear_model import SGDClassifier, LogisticRegression
from scipy.optimize import minimize
from statsmodels.tools.tools import add_constant
from copy import deepcopy
from IPython.display import display, clear_output
```

```python
#read data
df = pd.read_pickle("./data/FK_Acceptance_Rates.pkl") 

#context i.e. features
X = (df.drop(['parm_u', 'parm_s'], axis = 1)).to_numpy()

parm_map = df.drop_duplicates()

feature_list = (df.drop(['parm_u', 'parm_s'], axis = 1)).columns.values.tolist()
```

```python
import SmP_Bandit as cmab
from SmP_CMAB_Sim import cmab_sim as sim

# instance of simulation class
simulation = sim(n_choices = 20,
                       bp_conv_fac = 0.5,
                       parameter_map = parm_map, 
                       feature_list = feature_list)

# instance of bandit class
bts = cmab.Smart_Bandit(algorithm = "thompson",
                       epsilon=0.1, 
                       n_boots = 20,
                       quantile = 0.8,
                       n_choices = 20,
                       base_model = LogisticRegression(fit_intercept=False),  
                       feature_list = feature_list,
                       reward_fct = simulation.get_real_reward,
                       seed= 1234)

egreedy = cmab.Smart_Bandit(algorithm = "epsilon_greedy",
                       epsilon=0.1, 
                       n_boots = 20,
                       quantile = 0.8,
                       n_choices = 20,
                       base_model = LogisticRegression(fit_intercept=False), 
                       feature_list = feature_list,
                       reward_fct = simulation.get_real_reward,
                       seed= 1234)

softmax = cmab.Smart_Bandit(algorithm = "softmax",
                       epsilon=0.1, 
                       n_boots = 20,
                       quantile = 0.8,
                       n_choices = 20,
                       base_model = LogisticRegression(fit_intercept=False), 
                       feature_list = feature_list,
                       reward_fct = simulation.get_real_reward,
                       seed= 1234)

ucb = cmab.Smart_Bandit(algorithm = "ucb",
                       epsilon=0.1, 
                       n_boots = 20,
                       quantile = 0.6,
                       n_choices = 20,
                       base_model = LogisticRegression(fit_intercept=False), 
                       feature_list = feature_list,
                       reward_fct = simulation.get_real_reward,
                       seed= 1234)

algorithms = [bts, egreedy, softmax, ucb] 
```

```python
# These lists will keep track of the rewards obtained by each policy
rewards_bts, rewards_egreedy, rewards_softmax, rewards_ucb = [list() for i in range(len(algorithms))]
lst_rewards = [rewards_bts, rewards_egreedy, rewards_softmax, rewards_ucb]

# batch size - algorithms will be refit after N rounds
batch_size = 10
inferred_oberservations = False 

# initial seed - all policies start with the same small random selection of actions/rewards
first_batch = X[:batch_size, :]
np.random.seed(4)
action_chosen = np.random.randint(simulation.n_choices, size=batch_size)
reactions_received, opt_arms = simulation.get_reactions(first_batch, action_chosen)

# maximize learning effect by creating all implied bids (less than and greater than; arm correlation)
if False: #inferred_oberservations:
    exp_X, exp_a, exp_r = algorithms[0].infer_observations(first_batch, action_chosen, reactions_received)
else:
    exp_X, exp_a, exp_r = first_batch, action_chosen, reactions_received

# fitting models for the first time
for algo in algorithms:
    algo.fit(X=exp_X, a=exp_a, r=exp_r)
    
# these lists will keep track of which actions does each policy choose
actions_hist_bts, actions_hist_egreedy, actions_hist_softmax, actions_hist_ucb = [exp_a.copy() for i in range(len(algorithms))]
lst_actions_hist = [actions_hist_bts, actions_hist_egreedy, actions_hist_softmax, actions_hist_ucb]

# These lists will keep track of the reactions obtained by each policy
reactions_hist_bts, reactions_hist_egreedy, reactions_hist_softmax, reactions_hist_ucb = [exp_r.copy() for i in range(len(algorithms))]
lst_reactions_hist = [reactions_hist_bts, reactions_hist_egreedy, reactions_hist_softmax, reactions_hist_ucb]

# These lists will keep track of the contexts (if observations are inferred they may differ)
X_hist = exp_X.copy() 
X_hist_bts, X_hist_egreedy, X_hist_softmax, X_hist_ucb = [exp_X.copy() for i in range(len(algorithms))]
lst_X_hist = [X_hist_bts, X_hist_egreedy, X_hist_softmax, X_hist_ucb]

# rounds are simulated from the full dataset
def simulate_rounds(algo_n, rewards, actions_hist, reaction_hist, X_hist, X_global,  
                    batch_st, batch_end, recency_weight):
    np.random.seed(batch_st)
    
    ## context 
    context_this_batch = X_global[batch_st:batch_end, :]
    ## choosing actions for this batch
    actions_this_batch = algorithms[algo_n].get_actions(X=context_this_batch, recency_weight=recency_weight)
    ## reaction
    reaction_this_batch, opt_arms_this_batch = simulation.get_reactions(context_this_batch, actions_this_batch)
        
    # maximize learning effect by creating all implied bids (less than and greater than; arm correlation)
    if inferred_oberservations:
        exp_X, exp_a, exp_r = algorithms[algo_n].infer_observations(context_this_batch, actions_this_batch, reaction_this_batch)
    else:
        exp_X, exp_a, exp_r = context_this_batch, actions_this_batch, reaction_this_batch
        
    # keeping track of the sum of real rewards received
    lst_rewards[algo_n].append(np.multiply(simulation.get_real_reward(actions_this_batch), reaction_this_batch).sum())
    # bernoulli rewards
    #lst_rewards[algo_n].append(reaction_this_batch.sum())
    
    # adding this batch to the history of selected actions
    new_actions_hist = np.append(actions_hist, exp_a)
    
    # adding this batch to the history of context
    new_X_hist = np.concatenate((X_hist, exp_X), axis=0)
    
    # adding this batch to the history of reactions
    new_reaction_hist = np.append(reaction_hist, exp_r)
    
    # now refitting the algorithms after observing these new rewards
    algorithms[algo_n].fit(X=new_X_hist, a=new_actions_hist, r=new_reaction_hist)
    
    return new_actions_hist, new_reaction_hist, new_X_hist

# now running all the simulation
for i in range(int(np.floor(X.shape[0] / batch_size))-1):
    clear_output(wait=True)
    display('Progress '+str("{0:.0%}".format(i/(np.floor(X.shape[0] / batch_size)))))
    batch_st = (i + 1) * batch_size
    batch_end = (i + 2) * batch_size
    batch_end = np.min([batch_end, X.shape[0]])
    recency_weight = np.ones(batch_size, dtype=np.int8)

    for algo_n in range(len(algorithms)):
        lst_actions_hist[algo_n], lst_reactions_hist[algo_n], lst_X_hist[algo_n]  = simulate_rounds(algo_n,
                                                                lst_rewards[algo_n],
                                                                lst_actions_hist[algo_n],
                                                                lst_reactions_hist[algo_n],
                                                                lst_X_hist[algo_n], 
                                                                X, 
                                                                batch_st, 
                                                                batch_end,
                                                                recency_weight)
```

```python
import matplotlib.pyplot as plt
from pylab import rcParams
%matplotlib inline

def get_mean_reward(reward_lst, batch_size=batch_size):
    mean_rew=list()
    for r in range(len(reward_lst)):
        mean_rew.append(sum(reward_lst[:r+1]) * 1.0 / ((r+1)*batch_size))
    return mean_rew

rcParams['figure.figsize'] = 25, 15
lwd = 5
cmap = plt.get_cmap('tab20')
colors=plt.cm.tab20(np.linspace(0, 1, 20))

fig = plt.figure()
ax = plt.subplot(111)
plt.plot(get_mean_reward(lst_rewards[0]), label="Bootstrap Thompson Sampling" ,linewidth=lwd,color=colors[0])
plt.plot(get_mean_reward(lst_rewards[1]), label="Epsilon Greedy" ,linewidth=lwd,color=colors[1])
plt.plot(get_mean_reward(lst_rewards[2]), label="Softmax" ,linewidth=lwd,color=colors[2])
plt.plot(get_mean_reward(lst_rewards[3]), label="Upper Confidence Bound" ,linewidth=lwd,color=colors[3])

# import warnings
box = ax.get_position()
ax.set_position([box.x0, box.y0 + box.height * 0.1,
                 box.width, box.height * 1.25])
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, ncol=3, prop={'size':20})

plt.tick_params(axis='both', which='major', labelsize=25)
plt.xticks(np.arange(0, len(lst_rewards[0]), step=len(lst_rewards[0])/10))


plt.xlabel('Rounds', size=30)
plt.ylabel('Cumulative Mean Reward', size=30)
plt.title('Comparison of CMAB Algorithms\n(Base Algorithm is Logistic Regression)',size=30)
plt.grid()
#plt.tight_layout()
plt.show()
fig.savefig("output/cmab_comp_plot.png", bbox_inches = "tight")
```
